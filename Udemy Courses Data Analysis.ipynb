### Graphic Design courses that are of price 100rs

###  Courses related to Python

```python
df[df['course_title'].str.contains('Python')]
```

### Begineer Level Python Courses


```python
df[df['level'].str.contains('Beginner Level') & (df['course_title'].str.contains('Python'))]
```

### courses published in 2017

```python
df['published_timestamp']=pd.to_datetime(df['published_timestamp'])
df.head(2)
```

```python
df.dtypes
```

```python
df['year']=df['published_timestamp'].dt.year
df.head(3)
```

```python
 df[df['year']==2017]
```

### Max Number of subscriber for each level of courses

```python
df.groupby('level')['num_subscribers'].max()
```

## Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(df,hue="subject" , palette='brg_r')
```

```python
plt.figure(figsize=(20,6))
sns.countplot(x='subject',data=df)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.barplot(x ="num_subscribers", y ="subject", data = df, 
            )
```

```python
sns.barplot(x ="num_subscribers", y ="level", data = df, 
            palette ='plasma')
```

```python
plt.figure(figsize=(20,6))
sns.swarmplot(x='subject', y='level', data=df)
```

```python
plt.figure(figsize=(20,6))
sns.countplot(x='level',data=df)
plt.show()
```
